```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("group1.txt", delimiter=r"\s+")
data.head()
```

```python
fig, ax = plt.subplots(nrows=4, sharex=True, sharey=True, figsize=(8, 12))
data.plot("depth[m]", "KN[Ohmm]", marker="o", ax=ax[0])
data.plot("depth[m]", "SB[Ohmm]", marker="o", ax=ax[1])
data.plot("depth[m]", "OK[Ohmm]", marker="o", ax=ax[2])
data.plot("depth[m]", "UK[Ohmm]", marker="o", ax=ax[3])
ax[0].set_yscale("log")
for a in ax: a.grid()
```
