```python
spec.sortData()
```

```python
fig, ax = spec.showDataKK()
# fig.savefig("pyritGrob-rawKK.svg")
```

```python
spec.removeEpsilonEffect()
spec.showAll();
```

```python
fig, ax = spec.showDataKK()
# fig.savefig("pyritGrob-corrKK.svg")
```

```python
spec.cutF(1000.0)
spec.cutF(0.1, down=True)
fig, ax = spec.showData()
# fig.savefig("filtered.svg")
```
